**Bichinho Virtual:**

Crie uma classe que modele um Bichinho Eletrônico:

**Atributos:** Nome, Fome, Saúde e Idade

**Métodos:** Alterar Nome, Fome, Saúde e Idade; Retornar Nome, Fome, Saúde e  Idade

**Obs:** Existe mais uma informação que devemos levar em consideração, o  Humor do nosso Bichinho Eletrônico, este humor é uma combinação entre os atributos Fome e Saúde, ou seja, um campo calculado, então não devemos criar um atributo para armazenar esta informação por que ela pode ser calculada a qualquer momento (e.g., fome * saúde).

```python
# Resolução
class bichoV:
    """
    Cria um bichinho virtual e permite alterar seus atributos
    """
    
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        self.fome = 0
        self.saude = 100
        
    def set_fome(self, nova_fome):
        '''
        Altera a fome do bichinho com valores de 0 a 100
        '''
        if type(nova_fome) not in (float, int):
            print('Insira um numero (0-100)')
            return None
        if (nova_fome < 0) or (nova_fome > 100):
            print('Insira um numero (0-100)')
            return None
        else:
            print(f'Fome alterada de {self.fome} para {nova_fome}')
            self.fome = nova_fome
    
    def set_nome(self, novo_nome):
        print(f'Nome alterado de {self.nome} para {novo_nome}')
        self.nome = novo_nome
        
    def set_saude(self, nova_saude):
        '''
        Altera a saude do bichinho com valores de 0 a 100
        '''
        if type(nova_saude) not in (float, int):
            print('Insira um numero (0-100)')
            return None
        if (nova_saude < 0) or (nova_saude > 100):
            print('Insira um numero (0-100)')
            return None
        else:
            print(f'Saude alterada de {self.saude} para {nova_saude}')
            self.saude = nova_saude
        
    def set_idade(self, nova_idade):
        '''
        Altera a idade do bichinho
        '''
        if type(nova_idade) not in (float, int):
            print('Insira um numero')
            return None
        else:
            print(f'Saude alterada de {self.idade} para {nova_idade}')
            self.idade = nova_idade
    
    def get_nome(self):
        '''
        Retorna o nome do bichinho
        '''
        return self.nome
    
    def get_fome(self):
        '''
        Retorna o fome do bichinho
        '''
        return self.fome
    
    def get_saude(self):
        '''
        Retorna o saude do bichinho
        '''
        return self.saude
    
    def get_idade(self):
        '''
        Retorna o idade do bichinho
        '''
        return self.idade
    
    def get_humor(self):
        self.humor = (self.saude-self.fome)*(self.idade/100)
        return self.humor
```

```python
# Resolução
class car:
    '''
    Recebe modelo do carro e consumo, fornece funcoes para andar, obter e adicionar combustivel.
    '''

    def __init__(self, model, fuel_usage, max_fuel):
        self.fuel = 0
        self.model = model

        if type(max_fuel) not in (float, int):
            raise TypeError('Maximum fuel must be a number')
        else:
            self.max_fuel = max_fuel
        
        if type(fuel_usage) not in (float, int):
            raise TypeError('Fuel usage must be a number')
        else:
            self.fuel_usage = fuel_usage

    
    def walk(self, distance):
        '''
        Anda com o carro uma distancia
        '''
        if type(distance) not in (float, int):
            raise TypeError('Distance must be a number')

        fuel_used = distance/self.fuel_usage
        if fuel_used > self.fuel:
            print(f'Sem combustivel suficiente\nDisponivel: {self.fuel}; Necessario: {fuel_used}')
            return False
        else:
            self.fuel -= fuel_used
            print(f'Andou {distance}km\nCombustivel utilizado: {fuel_used}; Disponivel: {self.fuel}')
            return True
    
    def get_fuel(self):
        return f'{self.fuel}/{self.max_fuel}'
    
    def add_fuel(self, fuel_add):
        if type(fuel_add) not in (float, int):
            raise TypeError('Distance must be a number')
        if self.fuel + fuel_add <= self.max_fuel:
            self.fuel += fuel_add
            print(f'{fuel_add}l de combustivel adicionado; Total do tanque: {self.fuel}')
            return True
        else:
            print(f'Combustivel excede o total no tanque: {self.fuel}/{self.max_fuel}')
            return False
            
        
    
```

**Conta Corrente:**

Crie  uma  classe  para  implementar  uma  conta  corrente.  

**A  classe  deve  possuir  os  seguintes atributos:** número da conta, nome do correntista e saldo. 

**Os métodos são os seguintes:** alterarNome, depósito, saque, saldo.

No construtor, saldo é opcional, com valor default zero e os demais atributos são obrigatórios.

```python
# Resolução
from time import time
class cc:
    '''
    Classe para emulacao de conta corrente
    Contem nome, saldo e numero
    Pode alterar nome, depositar, sacar, e ver saldo
    '''
    def __init__(self, name, balance=0):
        self.acc_number = f'{time()}-{round(time()*0.99, 0)}'
        self.acc_holder = name  # nome
        self.balance = balance  # saldo
    
    def changeName(self, new_name):
        ''' Altera o nome da conta'''
        self.acc_holder = new_name
        return new_name
    
    def deposit(self, amount):
        ''' Deposita um valor na conta'''
        if amount < 0:
            raise ValueError('Amount for deposit must be greater than 0')
        else:
            print(f'Old balance: {self.balance}\nNew balance: {self.balance+amount}')
            self.balance+=amount
    
    def withdraw(self, amount):
        ''' Saca um valor da conta'''
        if amount < 0:
            raise ValueError('Amount for withdraw must be greater than 0')
        elif amount > self.balance:
            raise ValueError(f'Amount {amount} greater than account balance {self.balance}')
        else:
            self.balance-=amount
            print(f'Withdrew {amount}\nNew balance {self.balance}')
    
    def get_balance(self):
        ''' Retorna o saldo da conta'''
        print(f'Account balance: {self.balance}')
        return self.balance
    
    def get_account_info(self):
        ''' Retorna informacoes da conta'''
        print(f'Account number: {self.acc_number}\nAccount holder: {self.acc_holder}\nBalance: {self.balance}')
```

```python
conta = cc('Joao')
conta.get_account_info()
conta.deposit(100)
conta.withdraw(50)
conta.changeName('Adalberto')
conta.get_balance()
conta.get_account_info()
```

**Estatística Descritiva:**

Criar uma classe para implementar medidas descritivas.

Entre elas: média, mediana, valor máximo, valor mínimo, moda, desvio padrão, variância, amplitude, coeficiente de variação.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
dataset = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns= np.append(cancer['feature_names'], ['target']))
dataset
```

```python
type(dataset['mean radius'])
```

```python
# Resolução
from math import sqrt
class stat2:
    '''
    Recebe uma lista e possibilita o calculo de medidas estatisticas descritivas
    '''

    def __init__(self, data=[]):
        if type(data) == pd.core.series.Series:
            self.data_array = data
            self.data = data.tolist()
        else:
            self.data = data
        self.sorted_data = sorted(data)
        
    def mean(self):
        ''' Retorna a media'''
        return sum(self.data)/len(self.data)

    def median(self):
        ''' Retorna a mediana'''
        l_data = len(self.sorted_data)
        if l_data % 2 == 0:
            return (self.sorted_data[(l_data//2)-1] + self.sorted_data[l_data//2])/2
        else:
            return self.data[l_data//2]
        
    def Max(self):
        ''' Retorna o valor maximo'''
        return max(self.sorted_data)
    
    def Min(self):
        ''' Retorna o valor minimo'''
        return min(self.sorted_data)
    
    def mode(self):
        ''' Retorna a moda, metodo cru, ineficiente, mas sem usar funcoes prontas'''
        max_n = ('', 0)
        for element in self.data:
            n = self.data.count(element)
            if n > max_n[1]:
                max_n = (element, n)
        return max_n[0]
    
    def std_dev(self):
        ''' Retorna o desvio padrao'''
        x_ = self.mean()
        x = sum((self.data_array-x_)**2)
        return sqrt(x/(len(self.data)-1))
    
    def variance(self):
        ''' Retorna a variancia '''
        return self.std_dev()**2
    
    def amplitude(self):
        return self.Max()-self.Min()
    
    def coef_var(self):
        ''' Retorna o coeficiente de variacao '''
        return self.std_dev()/self.mean()
        
    
    
# x = [i for i in np.linspace(0, 2000, num=100)] + [2000]
c = stat2(dataset['mean radius'])
print(c.mean())
print(c.median())
print(c.Max())
print(c.Min())
print(c.mode())
print(c.std_dev())
print(c.variance())
print(c.amplitude())
print(c.coef_var())
```
